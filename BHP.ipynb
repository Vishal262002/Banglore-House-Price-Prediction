```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
file_path = '/content/drive/MyDrive/Analytics_Data /Bengaluru_House_Data.csv'
df1 = pd.read_csv(file_path)
df1.head()
```

```python
df1.shape
```

```python
df1.groupby('area_type')["area_type"].agg('count') # checking number of varities present in area_type column using groupby
```

```python
df2 = df1.drop(['area_type','society','balcony','availability'],axis='columns')  # removing the least important columns
df2.head()
```

```python
df2.isnull().sum() #Tells the count in particular columns that have 'NA' values in it using isnull()
```

```python
df3 = df2.dropna() # droping the NA values since they are in less number if compared with whole data
df3.isnull().sum()
```

```python
df3['size'].unique() #Checking all the possible unique values present in the size column
```

```python
df3['bhk'] = df3['size'].apply(lambda x: int(x.split(" ")[0])) #performing operation on size column and creating the new column to store the values of the operations performed
df3.head()
```

```python
df3['bhk'].unique()
```

```python
df3[df3.bhk>20] # showing values greater then 20 in bhk column
```

```python
df3['total_sqft'].unique()
```

```python
def is_float(x):
  try:
    float(x)
    return True
  except ValueError:
    return False
```

```python
df3[~df3['total_sqft'].apply(is_float)].head()

```

```python
def convert_sqft_to_num(x):
  tokens = x.split('-')
  if len(tokens) == 2:
      return (float(tokens[0]) + float(tokens[1])) / 2
  try:
      return float(x)
  except ValueError:
      return None
```

```python
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_to_num)
df4.head()
```

```python
df4.head(5)
```

```python
df5 = df4.copy()  #Finding per sqft are by multiplying price*1Lakhs / total_sqft
df5['price_per_sqft'] = df4['price']*100000/df4['total_sqft']
df5.head()
```

```python
len(df5.location.unique())
```

```python
df5.location = df5.location.apply(lambda x: x.strip()) #removing the extra spaces from the location column

location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending=False) #aperforming group by function on location column
location_stats
```

```python
len(location_stats[location_stats<=10]) #finding len of location column who are having data points less then or equal to 10 in location column
```

```python
location_stats_less_than_10 = location_stats[location_stats<=10] # assigning the location having <= 10 data poits in location_stast_less_than_10 variable
location_stats_less_than_10
```

```python
len(df5.location.unique())
```

```python
df5.location = df5.location.apply(lambda x: 'other' if x in location_stats_less_than_10  else x) # assigning location that are having data points <=10 to  amed as others
len(df5.location.unique())
```

```python
df5.head(10)
```

```python
df5[df5['total_sqft']/df5['bhk']<300].head()  #Dividing total sqft column by BHK
```

```python
df5.shape
```

```python
df6 = df5[~(df5.total_sqft/df5.bhk<300)] # removing outliers or data errors
df6.shape
```

```python
def remove_pps_outliers(df):
  df_out = pd.DataFrame()
  for key, subdf in df.groupby('location'):
    m = np.mean(subdf.price_per_sqft)
    st = np.std(subdf.price_per_sqft)
    reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
    df_out = pd.concat([df_out,reduced_df],ignore_index=True)
  return df_out

df7 =  remove_pps_outliers(df6)
df7.shape
```

```python
def plot_scatter_chart(df, location):
  bh2 = df[(df.location==location) & (df.bhk==2)]
  bh3 = df[(df.location==location) & (df.bhk==3)]
  matplotlib.rcParams['figure.figsize'] = (10,5)
  plt.scatter(bh2.total_sqft,bh2.price,color = 'blue',label = '2 BHK',s = 50)
  plt.scatter(bh3.total_sqft,bh3.price,marker = '+',color = 'green',label = '3 BHK',s = 50)
  plt.xlabel('Total Square Feet Area')
  plt.ylabel('Price')
  plt.title(location)
  plt.legend()

plot_scatter_chart(df7, "Hebbal")
```

```python
def remove_bhk_outliers(df):
  exclude_indices = np.array([])
  for location, location_df in df.groupby('location'):
    bhk_stats = {}
    for bhk, bhk_df in location_df.groupby('bhk'):
      bhk_stats[bhk] = {
          'mean': np.mean(bhk_df.price_per_sqft),
          'std': np.mean(bhk_df.price_per_sqft),
          'count' : bhk_df.shape[0]
      }
    for bhk, bhk_df in location_df.groupby('bhk'):
      stats = bhk_stats.get(bhk-1)
      if stats and stats['count']>5:
        exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
  return df.drop(exclude_indices,axis='index')

df8 = remove_bhk_outliers(df7)
df8.shape
```

```python
df9 = df8[df8.bath<df8.bhk+2]
df9.shape
```

```python
df10 = df9.drop(['size','price_per_sqft'],axis='columns')
df10.head(3)
```

```python
dummies = pd.get_dummies(df10.location)
dummies.head(3)
```

```python
df11 = pd.concat([df10,dummies.drop('other',axis='columns')],axis='columns')
df11.head(3)
```

```python
df12 = df11.drop('location',axis='columns')
df12.head(3)
```

```python
df12.shape
```

```python
X = df12.drop('price',axis='columns') #containing all independet columns and their valuesmb
X.head()
```

```python
y = df12.price
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits = 5,test_size = 0.2, random_state=0 )
cross_val_score(LinearRegression(),X,y,cv = cv)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
  algos = {
      'linear_regression' : {
          'model': LinearRegression(),
          'params': {
              'fit_intercept': [True,False]
          }
      },
      'lasso': {
          'model': Lasso(),
          'params': {
              'alpha': [1,2],
              'selection': ['random','cyclic']
          }
      },
      'decision_tree': {
          'model': DecisionTreeRegressor(),
          'params': {
              'criterion' :['mse','friedman_mse'],
              'splitter': ['best','random']
          }

      }
  }

  scores = []
  cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
  for algo_name, config in algos.items():
    gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
    gs.fit(X,y)
    scores.append({
        'model': algo_name,
        'best_score': gs.best_score_,
        'best_params': gs.best_params_
    })

  return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```

```python
def predict_price(location,sqft,bath,bhk):
  loc_index = np.where(X.columns==location)[0][0]

  x = np.zeros(len(X.columns))
  x[0] = sqft
  x[1] = bath
  x[2] = bhk
  if loc_index >= 0:
    x[loc_index] = 1

  return lr_clf.predict([x])[0]
```
